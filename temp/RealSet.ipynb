```python
f(x)=-2*(x^2+1)*(x-1)^2*(x+sqrt(2))
yf=1

rr=f.roots(multiplicities=true)
real_roots=[]
real_multi=[]
for i in range(len(rr)):
    if rr[i][0].is_real():
        real_roots.append(rr[i][0])
        real_multi.append(rr[i][1])
    
inds=sorted(range(len(real_roots)), key=lambda k: real_roots[k])
sort_roots=[]
sort_multi=[]
for i in range(len(inds)):
    sort_roots.append(real_roots[inds[i]])
    sort_multi.append(real_multi[inds[i]])

gap_roots=sort_roots[-1]-sort_roots[0]
if gap_roots>0:
    xi=sort_roots[0]-0.1*gap_roots
    xf=sort_roots[-1]+0.1*gap_roots
else:
    xi=sort_roots[0]-10
    xf=sort_roots[0]+10

xval=[xi]+sort_roots+[xf]

    
if f((xval[0]+xval[1])/2).n()>0:
    linef=line((xval[0],yf),(xval[1],yf))
else:
    linef=line((xval[0],yf),(xval[1],yf),linestyle='--')
    
for i in range(len(xval)-2):
    if f((xval[i+2]+xval[i+1])/2).n()>0:
        show((xval[i+1],yf))
        linef1=line((xval[i+1],yf),(xval[i+2],yf))
    else:
        linef1=line((xval[i+1],yf),(xval[i+2],yf),linestyle='--')
    pp=point((xval[i+1],yf),size=20)
    linef=linef+linf1+pp
```
