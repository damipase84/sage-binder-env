Il seguente codice crea una funzione per avere funzioni $x^{p/q}$ ben definite anche per $x<0$ quando $q$ è dispari:

```python
x=var('x')
def power_pq(x,p,q):
    if is_odd(q):
        val=((sign(x))^p)*abs(x)^(p/q)
    else:
        val=x^(p/q)   
    return val
```

```python
f(x)=power_pq(x,1,4) # esempio di radice cubica
plot(f,[-10,10])
```

## Grafico di funzioni potenza con potenza reale

Facciamo il grafico di funzioni del tipo
$$y=x^{\alpha}$$
dove $x>0$ e $\alpha\in \mathbb{R}$

Usiamo uno slider (in questo caso la potenza assume solo valori razionali):

```python
@interact
def _(a=slider(-3,3,step_size=0.1)):
    show(plot(x^a,(x,0, 8),ymin=-1,ymax=10, color='red',figsize=[6,4]))
```
