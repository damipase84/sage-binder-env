# Fattorizzazione di polinomi e ricerca degli zeri

Consideriamo ancora la funzione polinomiale di secondo grado: 
$$f(x)=x^3+2x^2-3x$$

Vediamo come possiamo trovarne gli zeri o fattorizzarla.

```python
x=var('x')  
f(x)=x^3+2*x^2-3*x
```

## Comando roots

Il metodo `.roots` applicato ad un polinomio restituisce gli zeri di un polinomio con la loro molteplicità

```python
%display latex
f.roots()
```

Questo ci dice che $-3, 1, 0$ sono zeri di $f$, tutti con molteplicità 1.

Facciamo un altro esempio:

```python
g(x)=x^4-2*x^2
show('g(x)=',g(x))
g.roots()
```

Infatti $g(x)=x^4-2x^2=x^2(x^2-2)=x^2(x-\sqrt{2})(x+\sqrt{2})$.

Gli zeri sono $x=0$ con molteplicità 2 e $x=\pm\sqrt{2}$ con molteplicità 1


Vediamo meglio com'è fatto l'output di `roots`

```python
r=g.roots()
r
```

`r` è una lista, in questo caso di lunghezza 3:

```python
len(r) # il comando `len` restituisce la lunghezza della lista
```

Accediamo all'elemento i-esimo della lista con il comando `r[i]`.

```python
show(r[0]) #Il primo indice è 0
show(r[1])
show(r[2]) # una lista di lunghezza 3 ha indici 0,1,2
```

Ogni elemento ha il valore dello zero, seguito dalla molteplicità. Per accedere al valore degli zeri possiamo usare il seguente comando:

```python
r[2][1]
```

Verifichiamo che i valori trovati sono effettivamente zeri di $g$:

```python
g(r[2][0])
```

**Attenzione**: `roots` cerca le radici nel campo dei numeri complessi:

```python
h(x)=x^4-1
show('h(x)=',h(x))
h.roots()
```

Infatti $h(x)=x^4-1=(x^2-1)(x^2+1)=(x+1)(x-1)(x^2+1)$

Per cercare solo le radici reali, dobbiamo impostare l'opzione `ring=RR`, ma attenzione che restituisce la funzione numerica:

```python
h.roots(ring=RR)
```

## Comando solve

```python
s=solve(f) # soluzione di f(x)=0
s
```

```python
show(solve(g))
```

Anche con `solve` possiamo vedere le molteplicità con l'opzione `multiplicities=true`

```python
s=solve(g,multiplicities=true)
s
```

Accediamo al valore degli zeri con questo comando:

```python
s[0][0].right() # il metodo .rhs prende il membro di destra
```

**Attenzione**: anche `solve` risolve nel campo dei numeri complessi:

```python
solve(h(x)==0)   # f1(x)=x^4-4
```

Per ottenre solo le soluzioni reali, dobbiamo dare le seguenti opzioni:

```python
solve(h(x)==0,algorithm='sympy',domain='real')
```

## Comando factor

Il comando `factor` ci permette di fattorizzare un polinomio nel campo dei numeri razionali

```python
show('f(x)=',f(x))
factor(f(x))
```

Facciamo un altro esempio:

```python
h(x)=x^2-4/9
show('h(x)=',h(x))
factor(h)
```

**Attenzione**: `factor` fattorizza solo se i coefficienti sono razionali

```python
show('g(x)=',g(x))
factor(g(x))
```

La fattorizzazione completa di $g(x)$ in $\mathbb{R}$ sarebbe:
\begin{align*}
x^4-2x^2=x^2(x-\sqrt{2})(x+\sqrt{2})
\end{align*}
