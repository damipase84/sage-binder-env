## Funzione esponenziale

Usiamo **SageMath**  per fare il grafico di alcune funioni esponenziali del tipo $f(x)=a^x$

```python
a=2 #base
```

```python
x=var('x');
f(x)=a^x;  # funzione esponenziale
plot(f)
```

Il comando *@interact* con l'opzione *slider* permette di vedere come cambia il grafico per diversi valori della base $a$ usando uno slider

```python
@interact
def _(a=slider(1.01,3,step_size=0.05)):
    p1=plot(a^x,(x,-5, 5),ymin=-0.5,ymax=5, color='red',legend_label='base a')
    p2=plot((1/a)^x,(x,-5, 5),ymin=-0.5,ymax=5, color='green',legend_label='base 1/a')
    show(p1+p2,figsize=[6,4])
```

Attenzione che la scala conta!

```python
plot(0.99^x,figsize=[6,3])
```

```python
plot(0.99^x,ymin=-1,ymax=10,figsize=[6,3])
```

```python
plot(0.99^x,(x,-500,500),ymin=-1,ymax=10,figsize=[6,3])
```

## Disequazioni esponenziali

Risolvere: $$ \frac{35}2\left(\frac15\right)^{2x} > \frac{7}{10} \cdot 5^x $$

Svolgimento
1) Condizioni di esistenza: $x\in \mathbb{R}$
2) Semplifichiamo 
\begin{align*}
 \frac{35}2\left(\frac15\right)^{2x} &> \frac{7}{10} \cdot 5^x \\
\frac{35}{7}5^{-2x}&>\frac{2}{10}5^x \\
5\cdot 5^{-2x} &>\frac{1}{5}\cdot 5^{x}\\
5^{-2x+1}&>5^{x-1}
\end{align*}
Siccome la base è $5>1$ possiamo passare alla disuguaglianza tra gli esponenti, senza cambiare il verso:
$$-2x+1>x-1\iff x<\frac 23$$

Risolviamo ora con Sage

```python
%display latex
dis=35/2*(1/5)^(2*x)>7/10* 5^x
dis
```

```python
solve(dis,algorithm='sympy')
```

Riproduciamo i passaggi: per prima cosa semplifichiamo i coefficienti

```python
dis1=dis/7*2
dis1
```

Usiamo le proprietà delle potenze:

```python
dis2=dis1.factor()
dis2
```

Siccome la base è 5>1, passiamo alla disuguaglianza agli esponenti e risolviamo:

```python
dis3=-2*x+1>x-1
dis3
```

```python
dis4=dis3+2*x+1
dis4/3
```
