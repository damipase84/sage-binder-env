## Funzione esponenziale

Usiamo **SageMath**  per fare il grafico di alcune funioni esponenziali del tipo $f(x)=a^x$

Per prima cosa definiamo la base $a$ e assegnamo il valore 2.

```python
a=2
```

Definiamo la funzione $f(x)=a^x$.

```python
x=var('x');
f(x)=a^x;
```

Valutiamo la funzione nel valore desiderato di $x$, ad esempio 3:

```python
f(3)
```

Valutiamo la funzione in diversi punti usando creando una lista con i punti desiderati ad esempio con il comando *range*:

```python
xval=range(-4,5);
list(xval)
```

```python
yval=[f(i) for i in xval]
list(yval)
```

## Grafico della funzione



Riproduciamo alcuni dei grafici presentati nelle slides.

Rappresentiamo i punti trovati nel piano cartesiano con il comando *list_plot*

```python
punti = list(zip(xval, yval))
list_plot(punti)
```

```python

# Plot the data
p_plot = list_plot(punti, color='red', marker='o',size=20)

# Display the plot
show(p_plot, axes_labels=['$x$', '$y$'])
```


Confrontiamo il grafico della funzione $y=2^x$ e la funzione $y=\left(\frac13\right)^x$

```python
g_plot=plot((1/3)^x,[-5,5],color='green',legend_label='$y=(1/3)^x$')
show(g_plot+f_plot, axes_labels=['$x$', '$y$'])
```

Il comando *@interact* con l'opzione *slider* permette di vedere come cambia il grafico per diversi valori della base $a$

```python
@interact
def _(a=slider(0.1,10,step_size=0.1)):
    show(plot(a^x,(x,-3, 3),ymin=-1,ymax=5, color='red'))
```
