# Equazioni polinomiali e razionali fratte

### Esempio - Equazione razionale fratta

Risolvere:

$$\dfrac{x^2-2x-4}{x^2+3x+2}=-\dfrac{1}{x+1}+\dfrac{1}{x+2}$$

**Soluzione**:
1) Condizioni di esistenza:
$$x^2+3x+2\neq 0\quad,\quad x+1\neq 0\quad ,\quad x+2\neq 0\ \iff \ x\neq -1\ ,\ x\neq -2$$
2) Semplifichiamo:
\begin{align*}
\dfrac{x^2-2x-4}{(x+2)(x+1)}&=\dfrac{(x+1)-(x+2)}{(x+2)(x+1)}\\
\dfrac{x^2-2x-3}{(x+2)(x+1)}&=0\\
\end{align*}
3) Soluzioni: $x^2-2x-3=(x+1)(x-3)$, per cui gli zeri del numeratore sono $x=-1$ e $x=3$.
Per le condizioni di esistenza, solo $x=3$ è accettabile.

Verifichiamo ora i passaggi usando Sage:

```python
%display latex
eqn=(x^2-2*x-4)/(x^2+3*x+2)==-1/(x+1)+1/(x+2)
eqn
```

`solve` ci dà direttamente la soluzione:

```python
solve(eqn,algorithm='sympy')
```

Verifica:

```python
eqn.substitute(x=3)
```

Usiamo Sage per fare i passaggi. Iniziamo con le condizioni di esistenza:

```python
show(eqn)
```

```python
solve([x^2+3*x+2!=0,x+2!=0,x+1!=0])
```

Ho preparato delle funzioni utili nel file `utils_DP.sage` .
Una di queste funzioni, `interval_sol`, semplifica l'output di `solve` scrivendolo come intervalli:

```python
load('utils_DP.sage') # comando per caricare le funzioni
ss=solve([x^2+3*x+2!=0,x+2!=0,x+1!=0])
interval_sol(ss)
```

Questo equivale a $x\neq -2$, $x\neq -1$. Da notare che la notazione usa le parentesi tonde per dire che l'estremo è escluso.

```python
show(eqn)
eqn1=eqn.factor()       # fattorizziamo entrambi i membri 
show(eqn1)
```

```python
eqn2=eqn1*(x+2)*(x+1)
show(eqn2)
```

```python
eqn3=eqn2+1
show(eqn3)
```

```python
solve(eqn3)
```

Chiaramente la soluzione $x=-1$ va esclusa. Possiamo informare `solve` che ci sono delle restrizioni sui valori possibili di $x$ mettendo a sistema con le condizioni di esistenza:

```python
solve([eqn3,x!=-1,x!=-2])
```

## Esempi visti durante le videolezioni

### Equazione razionale fratta

Risolvere:

$$\dfrac{x^2+x}{x^2-9}=\dfrac{1}{x+3}+\dfrac{2}{x-3}$$



```python
eqn=(x^2+x)/(x^2-9)==1/(x+3)+2/(x-3)
```

```python
solve(eqn)
```

```python
eqn.substitute(x=-1) # verifica
```

Ricreiamo i passaggi fatti per risolvere l'equazione. Partiamo dalle condizioni di esistenza


```python
ss=solve([x^2-9!=0,x+3!=0,x-3!=0])
interval_sol(ss)
```

Questi sono i valori di $x$ in cui cercare le soluzioni. Equivale a: $x\neq -3$, $x\neq 3$.

Vediamo come possiamo verificare i conti con Sage

```python
show(eqn)
eqn1=eqn.factor()       # fattorizziamo entrambi i membri
show(eqn1)
```

```python
eqn1=eqn1*(x-3)*(x+3)   # semplifichiamo i denominatori (diversi da zero)
show(eqn1)             
```

```python
eqn1=eqn1.expand()      # facciamo le moltiplicazioni
show(eqn1)
```

```python
eqn1=eqn1-eqn1.right()
show(eqn1)
```

```python
solve(eqn1)
```

Chiaramente $x=3$ va esclusa per le condizioni di esistenza:

```python
solve([eqn1,x!=-3,x!=3])
```
