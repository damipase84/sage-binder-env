# Polinomi in SageMath

## Definizione di una funzione

Iniziamo definendo la seguente funzione polinomiale di secondo grado: 
$$f(x)=x^2+2x-3$$


```python
x=var('x')  # questo comando definisce x come variabile simbolica
f(x)=x^3+2*x^2-3*x
```

In questa maniera abbiamo definito la funzione $f$ nella variabile $x$. Sage considera già la lettera `x` come una variabile simbolica, ma meglio ripeterlo.

Per vedere cos'è `f` per Sage usiamo i seguenti due comandi:

```python
f
```

Questo ci dice che Sage considera `f` come un'espressione simbolica, in questo caso definita proprio come la funzione polinomiale $f(x)=x^3+2x^2-3x$ 

Possiamo ottenere un formato in output più elegante con il comando `%display latex`

```python
%display latex
f
```

Per tornare al formato originale usiamo `%display play`:

```python
%display plain
f
```

Valutiamo $f$ per alcuni valori di $x$, ad esempio, $x=0$

```python
f(0)
```

Una maniera veloce per valutare una funzione su un elenco di valori è quello di creare una lista con i valori che ci interessano e usare il comando `map`.

Lista: $0, -0.5,-1/2,\sqrt{2},-3,-2$

`map` calcola: $f(0), f(-0.5),f(-1/2),\dots$

```python
xval=[0,-0.5,-1/2,sqrt(2),-2,-3,2]   #questo comando crea una lista
xval
```

```python
%display latex
yval=list(map(f,xval))  #list crea una lista sul risultato di map
yval
```

## Operazioni tra i polinomi

Consideriamo i due polinomi visti in classe e rivediamo le operazioni tra i polinomi in Sage

```python
p(x)=3*x^3+1/2*x+1
g(x)=-x^2+2*x
```

```python
p+g # addizione
```

```python
a=3
a*p # moltiplicazione per scalare
```

```python
p*g # moltiplicazione tra polinomi
```

```python
expand(p*g) # expand svolge le moltiplicazioni
```

Per fare la divisione polinomiale bisogna usare un metodo. Qui ad esempio dividiamo $P(x)$ per $D(x)$:

```python
P(x)=2*x^3-x+1
D(x)=-x^2+2*x
[Q,R]=P.maxima_methods().divide(D) #divisione con resto
show(Q)
show(R)
```

Facciamo la verifica che $D*Q+R=P$:

```python
D*Q+R
```

```python
expand(D*Q+R)
```

Questo va bene anche per verificare Ruffini: $R=0$ vuol dire che abbiamo una divisione esatta:

```python
D(x)=x+1
[Q,R]=P.maxima_methods().divide(D) #divisione con resto
show(Q)
show(R)
```

## Zeri del polinomio e fattorizzazione

### Comando roots

Il metodo `.roots` applicato ad un polinomio restituisce gli zeri di un polinomio con la loro molteplicità

```python
f.roots()
```

Questo ci dice che $-3, 1, 0$ sono zeri di $f$, tutti con molteplicità 1.

Facciamo un altro esempio:

```python
g(x)=x^4-2*x^2
g.roots()
```

Infatti la funzione $g(x)=x^4-x^2$ ha come zeri $x=0$ con molteplicità 2 e $x=\pm\sqrt{2}$ con molteplicità 1


Vediamo meglio com'è fatto l'output di `roots`

```python
r=f.roots()
r
```

`r` è una lista di lunghezza 3:

```python
len(r) # il comando `len` restituisce la lunghezza della lista
```

Accediamo all'elemento i-esimo della lista con il comando `r[i]`.

```python
show(r[0]) #Il primo indice è 0
show(r[1])
show(r[2])
```

Ogni elemento ha lo zero, seguito dalla molteplicità. Per accedere al valore degli zeri possiamo usare il seguente comando:

```python
show(r[1][0])
```

Verifichiamo che i valori trovati sono effettivamente zeri di $f$:

```python
f(r[1][0])
```

**Attenzione**: `roots` cerca le radici nel campo dei numeri complessi:

```python
f1(x)=x^4-4
f1.roots()
```

Per cercare solo le radici reali, dobbiamo impostare l'opzione `ring=RR`:

```python
f1.roots(ring=RR)
```

### Comando solve

Il comando `solve` può risolve equazioni e disequazioni. Quando diamo in argomento una funzione $f(x)$, `solve` cerca le soluzioni di $f(x)=0$ 

```python
s=solve(f)
s
```

```python
show(solve(g))
```

Anche con `solve` possiamo vedere le molteplicità con l'opzione `multiplicities=true`

```python
s=solve(g,multiplicities=true)
s
```

Accediamo al valore degli zeri con questo comando:

```python
s[0][0].right() # il metodo .rhs prende il membro di destra
```

**Attenzione**: anche `solve` risolve nel campo dei numeri complessi:

```python
solve(f1(x)==0)   # f1(x)=x^4-4
```

Per ottenre solo le soluzioni reali, dobbiamo dare le seguenti opzioni:

```python
solve(f1(x)==0,algorithm='sympy',domain='real')
```

### Comando factor

Il comando `factor` ci permette di fattorizzare un polinomio nel campo dei numeri razionali

```python
factor(f(x))
```

Facciamo un altro esempio:

```python
h(x)=x^2-4/9
factor(h)
```

**Attenzione**: `factor` fattorizza solo se i coefficienti sono razionali

```python
factor(g(x))
```

La fattorizzazione completa di $g(x)$ sarebbe:
\begin{align*}
x^4-2x^2=x^2(x-\sqrt{2})(x+\sqrt{2})
\end{align*}

## Grafico di una funzione

```python
plot(f) 
```

Il grafico viene creato di defoult per $x\in [-1,1]$. Cambiamo gli estremi usando `xmin` e `xmax`. Cambiamo colore, inseriamo la legenda e battezziamo gli assi.

```python
plot_f=plot(f,xmin=-4,xmax=3,color='red',legend_label=f(x),axes_labels=['$x$', '$y$'])
show(plot_f)
```

Mostriamo sul grafico i punti calcolati. Possiamo usare il comando `list_plot` per fare il grafico di punti. Ecco un esempio:

```python
list_plot([[0,1],[1,-1],[-1,3],[-1,2]],size=20)
```

Prima di usare `list_plot` sui punti trovati `xval` e `yval`, dobbiamo riorganizzarli in modo da avere una lista del tipo $[(x_1,y_1),(x_2,y_2),\dots]$. Si usa il comando `zip`

```python
punti = list(zip(xval, yval))  
plot_p=list_plot(punti,size=20)
show(plot_p+plot_f)
```
