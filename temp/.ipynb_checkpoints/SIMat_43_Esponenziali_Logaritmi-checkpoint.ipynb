## Funzione esponenziale

Usiamo **SageMath**  per fare il grafico di alcune funioni esponenziali del tipo $f(x)=a^x$

```python
a=2 #base
```

```python
x=var('x');
f(x)=a^x;  # funzione esponenziale
```

Il comando *@interact* con l'opzione *slider* permette di vedere come cambia il grafico per diversi valori della base $a$ usando uno slider

```python
@interact
def _(a=slider(1.01,3,step_size=0.01)):
    p1=plot(a^x,(x,-3, 3),ymin=-1,ymax=5, color='red',legend_label='base a')
    p2=plot((1/a)^x,(x,-3, 3),ymin=-1,ymax=5, color='green',legend_label='base 1/a')
    show(p1+p2)
```

Attenzione che la scala conta!

```python
plot(0.99^x)
```

```python
plot(0.99^x,ymin=-1,ymax=10)
```

```python
plot(0.99^x,(x,-500,500),ymin=-1,ymax=10)
```

## Disequazioni esponenziali

Risolvere: $$ \frac{35}2\left(\frac15\right)^{2x} > \frac{7}{10} \cdot 5^x $$

```python
%display latex
dis=35/2*(1/5)^(2*x)>7/10* 5^x
dis
```

```python
solve(dis,algorithm='sympy')
```

Riproduciamo i passaggi: per prima cosa semplifichiamo i coefficienti

```python
dis1=dis/7*2
dis1
```

Usiamo le proprietà delle potenze:

```python
dis2=dis1.factor()
dis2
```

Siccome la base è 5>1, passiamo alla disuguaglianza agli esponenti e risolviamo:

```python
dis3=-2*x+1>x-1
dis3
```

```python
dis4=dis3+2*x+1
dis4/3
```

## Funzioni logaritmiche
Vediamo la stessa figura per le funzioni logaritmiche.

Il logaritmo $\log_a(x)$ si esprime con la funzione `log(x,a)` : 

```python
log(4,2) 
```

```python
log(1/3,3)
```

Il logaritmo in base naturale, $\ln(x)$ si scrive `log(x)`

```python
log(sqrt(e))
```

```python
g(x)=log(x,2)  #funzione logaritmo in base 2
h(x)=log(x,1/2) #logaritmo in base 1/2
```

```python
@interact
def _(a=slider(1.01,3,step_size=0.01)):
    p1=plot(log(x,a),(x,0, 10),ymin=-5,ymax=5, color='red',legend_label='base $a$')
    p2=plot(log(x,1/a),(x,0, 10),ymin=-5,ymax=5, color='green',legend_label='base $1/a$')
    show(p1+p2)
```

## Disequazioni logaritmiche

Risolvere $$\ln(x+5) - \ln(4-x) + \ln(3x-1) > \ln(3x - 1) - \ln(x+4) $$

```python
dis=log(x+5) - log(4-x) + log(3*x-1) > log(3*x - 1) - log(x+4)
dis
```

```python
solve(dis,algorithm='sympy')
```

Risolviamo facendo tutti i passaggi. Per prima cosa calcoliamo i campi di esistenza:

```python
CE1=solve(3*x-1>0)
CE2=solve(x+5>0)
CE3=solve(-x+4>0)
CE4=solve(x+4>0)
CE1+CE2+CE3+CE4
```

Questo equivale a $\frac13 <x<4$.

Ora semplifichiamo cancellando i termini uguali e usando le proprietà dei logaritmi:

```python
dis1=dis-log(3*x-1)+log(x+4)
show(dis1)
dis1.log_simplify()
```

Ricordandoci che $0=\log(1)$, passiamo alla disuguaglianza tra gli argomenti, senza cambiare il verso, base $e> 1$

```python
dis2=-(x+5)*(x+4)/(x-4)>0
dis2
```

```python
solve(dis2)
```
