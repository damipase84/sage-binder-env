## Risoluzione di disequazioni irrazionali

### Esempio 2

Risolvere
$$\sqrt{2x-1}-\sqrt{x-4}>2$$

1) Condizioni di esistenza:
   \begin{align*}
   \begin{cases}
   2x-1\geq 0\\
   x-4\geq 0
   \end{cases}
   \iff
   \begin{cases}
   x\geq 1/2\\
   x\geq 4
   \end{cases}
   \iff x\geq 4
   \end{align*}
2) Semplifichiamo elevando a quadrato entrambi i membri. Possiamo farlo perché il membro di destra e' sempre positivo, mentre il membro di sinistra è non negativo per $2x-1\geq x-4\iff x\geq -3$ (sempre verificato nelle condizioni di esistenza):
    \begin{align*}
   \left(\sqrt{2x-1}-\sqrt{x-4}\right)^2 &> 2^2\\
   2x-1+x-4-2\sqrt{(2x-1)(x-4)} &> 4\\
     -2\sqrt{(2x-1)(x-4)} &> -3x +9\\
    2\sqrt{(2x-1)(x-4)} &< 3x -9\\
   \end{align*}
3) Date le condizioni di esistenza ($x\geq 4$), il termine di destra $g(x)=3x-9$ è sempre positivo e possiamo elevare al quadrato:
\begin{align*}
   4(2x-1)(x-4) &< (3x -9)^2\\
   4(2x^2-8x -1x+4) &< 9x^2-54x +81\\
   -x^2+18x -65 &< 0\\
   \end{align*}
   Gli zeri della quadratica sono $x_1=5$ e $x_2=13$, da cui $x<5 \ \lor x>13$.

   Tenendo conto delle CE, le soluzioni  sono $4\leq x < 5 \ \lor x>13$

   Vediamo le soluzioni con Sage:

```python
%display latex
eqn=sqrt(2*x-1)-sqrt(x-4)>2
eqn
```

```python
ss=solve(eqn,algorithm='sympy')
show(ss)
load('utils_DP.sage')
interval_sol(ss)
```

In questo caso Sage ottiene le soluzioni giuste. Riproduciamo i passaggi con Sage.
Per prima cosa le condizioni di esistenza:

```python
ce=solve([2*x-1>=0,x-4>=0])
interval_sol(ce)
```

```python
eqn1=eqn^2
show(eqn1)
```

```python
eqn2=eqn1.expand()
show(eqn2)
```

```python
eqn3=(eqn2-3*x+5)/2
show(eqn3)
```

```python
eqn4=eqn3-eqn3.left()-eqn3.right()
show(eqn4)
```

```python
g(x)=eqn4.left()
f(x)=(eqn4.right())^2
show('g(x)=',g(x))
show('f(x)=',f(x))
```

```python
solve(g(x)>=0)
```

```python
eqn5=(g(x))^2>f(x)
show(eqn5)
```

```python
eqn6=eqn5-eqn5.right()
eqn6=eqn6.expand()
eqn6
```

```python
solve([eqn6])
```

```python
ss=solve([x>=4,eqn6])
interval_sol(ss)
```

### Esempio - disequazione radicale con indice pari

```python
dis=sqrt(x^2-9)>= 3-1/5*x
dis
```

```python
solve(dis,algorithm='sympy')
```

Facciamo i passaggi.
Caso 1: rhs >=0
Risolviamo:

$$\text{Caso 1}\  \begin{cases}
-\frac15 x+3\geq 0 \\
x^2-9\geq 0\\
x^2-9\geq (-\frac 15 x+3)^2
\end{cases}\qquad,\qquad \text{Caso 2}\ \begin{cases}
-\frac15 x+3< 0 \\
x^2-9\geq 0\\
\end{cases} $$

```python
g(x)=dis.right()
g(x)
```

```python
f(x)=dis.left()^2
f(x)
```

```python
dis^2
```

```python
S1=solve([g(x)>=0,f(x)>=0,dis^2])
S1
```

```python
S2=solve([g(x)<0,f(x)>=0])
S1+S2
```

```python
interval_sol(S1+S2)
```
