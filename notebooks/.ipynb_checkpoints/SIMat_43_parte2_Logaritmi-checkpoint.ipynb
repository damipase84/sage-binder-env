Usiamo **SageMath**  per fare il grafico di alcune funioni esponenziali del tipo $f(x)=a^x$

```python
a=2 #base
```

```python
x=var('x');
f(x)=a^x;  # funzione esponenziale
plot(f)
```

Il comando *@interact* con l'opzione *slider* permette di vedere come cambia il grafico per diversi valori della base $a$ usando uno slider

```python
@interact
def _(a=slider(1.01,3,step_size=0.01)):
    p1=plot(a^x,(x,-5, 5),ymin=-0.5,ymax=5, color='red',legend_label='base a')
    p2=plot((1/a)^x,(x,-5, 5),ymin=-0.5,ymax=5, color='green',legend_label='base 1/a')
    show(p1+p2,figsize=[6,5])
```

Attenzione che la scala conta!

```python
plot(0.99^x)
```

```python
plot(0.99^x,ymin=-1,ymax=10)
```

```python
plot(0.99^x,(x,-500,500),ymin=-1,ymax=10)
```

Risolviamo ora con Sage

```python
%display latex
dis=35/2*(1/5)^(2*x)>7/10* 5^x
dis
```

```python
solve(dis,algorithm='sympy')
```

Riproduciamo i passaggi: per prima cosa semplifichiamo i coefficienti

```python
dis1=dis/7*2
dis1
```

Usiamo le proprietà delle potenze:

```python
dis2=dis1.factor()
dis2
```

Siccome la base è 5>1, passiamo alla disuguaglianza agli esponenti e risolviamo:

```python
dis3=-2*x+1>x-1
dis3
```

```python
dis4=dis3+2*x+1
dis4/3
```

## Disequazioni logaritmiche

Risolvere $$\ln(x+5) - \ln(4-x) + \ln(3x-1) > \ln(3x - 1) - \ln(x+4) $$

Svolgimento:

1) Condizioni di esistenza:
\begin{align*}
\begin{cases}
x+5>0\\
4-x>0\\
3x-1>0\\
x+4>0
\end{cases}
\iff
\begin{cases}
x>-5\\
x<4\\
x>1/3\\
x>-4
\end{cases}
\iff \frac 13 < x<4
\end{align*}

2) Semplifichiamo usando le proprietà dei logaritmi:
\begin{align*}
\ln(x+5)+\ln(x+4)&>\ln(4-x)\\
\ln((x+5)\cdot(x+4))&>\ln(4-x)
\end{align*}

3) Siccome la base è $e>1$, possiamo passare alla disuguaglianza tra gli argomenti, senza cambiare il verso:
\begin{align*}
&(x+5)\cdot(x+4)>4-x\\
&x^2+9x+20+x-4>0\\
&x^2+10x+16>0
\end{align*}

$\Delta=100-4\cdot 16=36\implies \ x_1=-8,\ x_2=-2$
Le soluzioni della funzione quadratica sono: $x<-8\ \lor\ x>-2$

4) Tenendo conto delle condizioni di esistenza abbiamo le soluzioni:
   $\frac 13 < x <4$
 

Vediamo ora come risolvere con Sage

```python
dis=log(x+5) - log(4-x) + log(3*x-1) > log(3*x - 1) - log(x+4)
dis
```

```python
solve(dis,algorithm='sympy')
```

Risolviamo facendo tutti i passaggi. Per prima cosa calcoliamo le condizioni di esistenza:

```python
solve([x+5>0,4-x>0,3*x-1>0,x+4>0])
```

Questo equivale a $\frac13 <x<4$.

Ora semplifichiamo cancellando i termini uguali e usando le proprietà dei logaritmi:

```python
dis1=dis-log(3*x-1)+log(x+4)
show(dis1)
dis1.log_simplify()
```

Ricordandoci che $0=\log(1)$, passiamo alla disuguaglianza tra gli argomenti, senza cambiare il verso, base $e> 1$

```python
dis2=-(x+5)*(x+4)/(x-4)>0
dis2
```

```python
solve(dis2)
```
