# Equazioni e disequazioni

Sage può risolvere equazioni e disequazioni. Per esempio ricaviamo la formula risolutiva per le equazioni quadratiche:

```python
%display latex
a,b,c = var('a,b,c')
f(x) = a*x^2 + b*x + c # la lettera x è automaticamente una variabile simbolica
sol=solve(f(x), x)
show(sol)
```

Possiamo valutare il valore degli zeri assegnando dei valori ai parametri:

```python
sol[0](a=1,b=3,c=2)
```

## Operatori di confronto

Abbiamo i seguenti simboli per confrontare due valori nelle uguaglianze e diseguaglianze:

`==`  corrisponde a $=$ nelle equazioni

`>=` corrisponde a $\geq$ delle disequazioni

`<=` corrisponde a $\leq$ 

`!=` corrisponde a $\neq$

Il risultato di un confronto è una variabile booleana, `True` o `False`

```python
3==2
```

```python
3==3
```

```python
3>=2
```

```python
3<=2
```

```python
3!=2
```

## Operazioni con equazioni e disequazioni

Vediamo come Sage ci può aiutare a risolvere equazioni o disequazioni.

Consideriamo la seguente equazione, che sappiamo risolvere facilmente:

\begin{align*}
x^2+2x = 2x +1\iff x^2-1=0\iff x=\pm1
\end{align*}

```python
eqn = x^2+2*x == 2*x +1  # la variabile eqn contiene l'equazione che ci interessa
eqn
```

```python
eqn.left()  # membro di sinistra
```

```python
eqn.right() #membro di destra
```

Il comando `solve` risolve l'equazione

```python
solve(eqn)
```

Possiamo fare la verifica sostituendo:

```python
eqn.substitute(x=0)
```

Vogliamo usare Sage per verificare i nostri passaggi. 
Possiamo ricreare i nostri conti facendo le usuali operazioni di addizione e moltiplicazione su `eqn`,  e vengono eseguite ad entrambi i membri. Ad esempio:

```python
show(eqn)
```

```python
eqn1=eqn-eqn.right() # sottraiamo ad entrambi i membri il membro di destra
show(eqn1)
```

```python
show(eqn)
eqn*2      # moltiplichiamo per 2 entrambi i membri
```

```python
eqn*(-1) # moltiplichiamo per -1 entrambi i membri
```

**ATTENZIONE** 
Le stesse operazioni si possono fare anche con `<`, `<=`, `>`, `>=`. La **moltiplicazione per numeri negativi** in una disequazione non cambia il verso!

```python
dis= -x>3
show(dis)
```

```python
dis*(-1)
```

Un'alternativa alla moltiplicazione per -1, e' sottrarre entrambi i membri:

```python
dis-dis.left()-dis.right()  # equivale a moltiplicare per -1
```
