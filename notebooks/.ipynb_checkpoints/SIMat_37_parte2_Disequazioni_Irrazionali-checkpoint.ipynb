# Disequazioni irrazionali - 1

## Risoluzione di disequazioni irrazionali

### Esempio 1

Risolvere $$1\leq x-2\sqrt{2x-3}$$

Passaggi:
1) Condizioni di esistenza: $f(x)=2x-3\geq 0\iff x\geq \frac 32$
2) Semplifichiamo:
\begin{align*}
1&\leq x-2\sqrt{2x-3}\\
2\sqrt{2x-3}&\leq x-1\\
\sqrt{2x-3}&\leq \frac{x}{2}-\frac12\\
\end{align*}
3) $g(x)=\frac{x}{2}-\frac12$.
   $$\text{Caso 1):}\  \begin{cases}
g(x)\geq 0\\
f(x)\geq 0\\
f(x)\leq g^2(x)
\end{cases}\quad ,\quad  \text{Caso 2):}\ \begin{cases}
g(x)< 0\\
f(x)\geq 0\\
\sqrt{f(x)}\leq g(x) 
\end{cases}
\implies \nexists $$

Caso 1: 
\begin{align*}
\begin{cases}
\frac{x}{2}-\frac12\geq 0\\
2x-3\geq 0\\
2x-3\leq \frac 14 (x^2-2x+1)
\end{cases}
\iff \begin{cases}
x\geq 1\\
x\geq 3/2\\
x^2-10x+13\geq 0
\end{cases}
\end{align*}

Troviamo gli zeri della funzione quadratica:
$\Delta=10^2-13\cdot 4=12\cdot4=3\cdot 4^2$

$x_1=\frac{10-4\sqrt{3}}{2}=5-2\sqrt{3}\approx 1,54 \quad,\quad x_2=\frac{10+4\sqrt{3}}{2}=5+2\sqrt{3}\approx 8,46 $

\begin{align*}
\begin{cases}
x\geq 1\\
x\geq 3/2\\
x\leq 5-2\sqrt{3} \ \lor x\geq 5+2\sqrt{3}
\end{cases} \iff \frac 32 \leq x\leq 5-2\sqrt{3} \quad \lor\quad  x\geq 5+2\sqrt{3}
\end{align*}

Vediamo la risoluzione in Sage

```python
%display latex
dis=1<=x-2*sqrt(2*x-3)
dis
```

Questo è un caso in cui `solve` non riesce a trovare le soluzioni esatte, neppure con l'opzione `algorithm='sympy'` 

```python
solve(dis)
```

```python
solve(dis,algorithm='sympy')
```

Vediamo i passaggi per risolvere questa disequazione irrazionale.

1) Condizioni di esistenza:

```python
show(dis)
solve(2*x-3>=0)
```

2) Semplifichiamo

```python
dis1=(dis-x)/2
dis1
```

Invece di moltiplicare per $-1$, invertiamo i membri:

```python
dis2=dis1-dis1.right()-dis1.left()
dis2
```

3) Dividiamo i casi:

 $$\text{Caso 1):}\  \begin{cases}
g(x)\geq 0\\
f(x)\geq 0\\
f(x)\leq g^2(x)
\end{cases}\quad ,\quad  \text{Caso 2):}\ \begin{cases}
g(x)< 0\\
f(x)\geq 0\\
\sqrt{f(x)}\leq g(x) 
\end{cases}
\implies \nexists $$

```python
f(x)=dis2.left()^2
g(x)=dis2.right()
solve(g(x)>=0) # condizione caso 1
```

```python
solve(f(x)>=0) # condizioni esistenza
```

```python
dis3=f(x)<=(g(x))^2
show(dis3)
```

```python
dis3=dis3.expand()
show(dis3)
```

```python
dis3=dis3-dis3.right()
dis3()
```

```python
solve(dis3) # soluzioni equazione di secondo grado
```
