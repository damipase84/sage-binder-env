## Disequazioni

### Esempio 1 - Disequazione polinomiale

Risolvere
$$t^6+8t^5\geq -12t^4$$

Soluzione:
\begin{align*}
t^6+8t^5+12t^4&\geq 0\\
t^4(t^2+8t+12) &\geq 0\\
t^4(t+6)(t+2)  &\geq 0 
\end{align*}
$t^4$ è un fattore non negativo, per cui il segno è dato da $(t+6)(t+2)$:
$$t\leq -6 \ \lor t\geq -2$$

Vediamo come ottenere queste soluzioni in Sage.

```python
%display latex
t=var('t')
dis=t^6+8*t^5>=-12*t^4
dis
```

```python
solve(dis)
```

Questo equivale a $t\leq -6 \lor t\geq -2$, infatti:

```python
load('utils_DP.sage') # carica le funzioni nel file utils_DP
ss=solve(dis)
interval_sol(ss)
```
