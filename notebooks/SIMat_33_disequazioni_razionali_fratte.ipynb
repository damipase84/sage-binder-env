# Equazioni e disequazioni razionali fratte

Sage può risolvere equazioni e disequazioni. Per esempio ricaviamo la formula risolutiva per le equazioni quadratiche:

```python
%display latex
a,b,c = var('a,b,c')
f(x) = a*x^2 + b*x + c # la lettera x è automaticamente una variabile simbolica 
show(solve(f(x), x))
```

## Operatori di confronto

Abbiamo i seguenti simboli per confrontare due valori nelle uguaglianze e diseguaglianze:

`==`  corrisponde a $=$ nelle equazioni

`>=` corrisponde a $\geq$ delle disequazioni

`<=` corrisponde a $\leq$ 

`!=` corrisponde a $\neq$

Il risultato di un confronto è una variabile booleana, `True` o `False`

```python
3==2
```

```python
3==3
```

```python
3>=2
```

```python
3<=2
```

```python
3!=2
```

## Equazioni

Vediamo come Sage ci può aiutare a risolvere equazioni o disequazioni.

Consideriamo la seguente equazione, che sappiamo risolvere facilmente:

\begin{align*}
x^2+2x = 2x +1\iff x^2-1=0\iff x=\pm1
\end{align*}

```python
eqn = x^2+2*x == 2*x +1  # la variabile eqn contiene l'equazione che ci interessa
eqn
```

```python
eqn.left()  # membro di sinistra
```

```python
eqn.right() #membro di destra
```

Il comando `solve` risolve l'equazione

```python
solve(eqn)
```

Possiamo fare la verifica sostituendo:

```python
eqn.substitute(x=1)
```

Vogliamo usare Sage per verificare i nostri passaggi. 
Possiamo ricreare i nostri conti facendo le usuali operazioni di addizione e moltiplicazione su `eqn`,  e vengono eseguite ad entrambi i membri. Ad esempio:

```python
eqn-2*x-1  
```

```python
eqn*2
```

```python
eqn*(-1)
```

### Esempio 1 - Equazione razionale fratta

Risolvere:

$$\dfrac{x+2}{x^2+3x+2}==-\dfrac{1}{x+1}+\dfrac{1}{x+2}$$


```python
eqn=(x^2-2*x-4)/(x^2+3*x+2)==-1/(x+1)+1/(x+2)
eqn
```

`solve` ci dà direttamente la soluzione:

```python
solve(eqn,algorithm='sympy')
```

verifichiamo:

```python
eqn.substitute(x=3)
```

Usiamo Sage per fare i passaggi:

```python
CE1=solve((x^2+3*x+2))
CE2=solve(x+2)
CE3=solve(x+1)
CE1+CE2+CE3
```

```python
show(eqn)
eqn1=eqn.factor()       # fattorizziamo entrambi i membri 
show(eqn1)
```

```python
eqn2=eqn1*(x+2)*(x+1)
show(eqn2)
```

```python
eqn3=eqn2+1
show(eqn3)
```

```python
solve(eqn3)
```

Chiaramente la soluzione $x=-1$ va esclusa. Possiamo informare `solve` che ci sono delle restrizioni sui valori possibili di $x$ con il comando `assume`

```python
assume(x!=-1,x!=-2)
solve(eqn3)
```

## Disequazioni

### Esempio 1 - Disequazione polinomiale

Risolvere
$$t^6+8t^5\geq -12t^4$$

```python
t=var('t')
dis=t^6+8*t^5>=-12*t^4
dis
```

```python
solve(dis)
```

Questo equivale a $t\leq -6 \lor t\geq -2$.

Sage ci può aiturare a verificare i nostri passaggi:

```python
dis1=dis-dis.right()
dis1
```

```python
dis1.left().factor()
```

A questo punto ci rimande da fare lo studio del segno

### Esempio 2 - Disequazione razionale fratta

```python
dis=(3*x+1)/(x+4)>=1
dis
```

```python
solve(dis)
```

Vediamo di verificare i passaggi con sage:

```python
assume(x!=-4)  # Condizioni di esistenza
```

```python
dis1=dis-1
show(dis1)
```

```python
dis1=dis1.factor()
show(dis1)
```

A questo punto bisogna fare lo studio del segno di numerotore e denominatore

**ATTENZIONE** 
la moltiplicazione per numeri negativi in una disequazione non cambia il verso!

```python
dis= -x>3
show(dis)
```

```python
dis1=dis*(-1)
show(dis1)
```

Un'alternativa alla moltiplicazione per -1, e' sottrarre entrambi i membri:

```python
dis2=dis-dis.left()-dis.right()  # equivale a moltiplicare per -1
dis2
```

## Esempi visti in classe

### Equazione razionale fratta

Risolvere:

$$\dfrac{x^2+x}{x^2-9}=\dfrac{1}{x+3}+\dfrac{2}{x-3}$$



```python
eqn=(x^2+x)/(x^2-9)==1/(x+3)+2/(x-3)
```

```python
solve(eqn)
```

```python
eqn.substitute(x=-1)
```

Ricreiamo i passaggi fatti per risolvere l'equazione. Partiamo dalle condizioni di esistenza


```python
CE1=solve((x^2-9))
CE2=solve(x+3)
CE3=solve(x-3)
CE1+CE2+CE3
```

Questi sono i valori di $x$ da escludere dalle soluzioni.

Vediamo come possiamo verificare i conti con Sage

```python
show(eqn)
eqn1=eqn.factor()       # fattorizziamo entrambi i membri
show(eqn1)
```

```python
eqn1=eqn1*(x-3)*(x+3)   # semplifichiamo i denominatori (diversi da zero)
show(eqn1)             
```

```python
eqn1=eqn1.expand()      # facciamo le moltiplicazioni
show(eqn1)
```

```python
eqn1=eqn1-eqn1.right()
show(eqn1)
```

```python
solve(eqn1)
```

```python
assume(x!=3,x!=-3)
solve(eqn1)
```

### Disequazione polinomiale 

```python
dis=x^3+4*x^2-12*x< 0
dis
```

```python
solve(dis)
```

Questo equivale a $x<-6 \lor 0<x<2$.

Sage ci può aiturare a verificare i nostri passaggi ad esempio fattorizzando il membro di sinistra:

```python
dis1=dis.left().factor()
dis1
```

```python
dis=2*(-x^2-1)*(x-1)^2*(x+sqrt(2))<= 0
dis
```

```python
solve(dis)
```

Questo equivale a $-\sqrt{2}<x<1 \ \cup \ x>1$
